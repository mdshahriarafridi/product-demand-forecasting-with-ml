```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('dataset/demand_forecasting.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe().transpose().round(2)
```

```python
df.describe(include='object').T
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df["Weekday"] = df['Date'].dt.weekday
```

```python
df.head()
```

```python
df.columns
```

```python
df["Discounted Price"] = df['Price']*(1 - df['Discount'] / 100)
```

```python
df.head()
```

```python
df["Sell Through Rate"] = df['Units Sold'] / df['Inventory Level']
```

```python
df["Sell Through Rate"].describe()
```

```python
df.groupby('Category')["Demand"].agg(["mean",'sum',"std",]).sort_values(by = 'sum', ascending=False).round(2)
```

```python
df.groupby(["Region", "Seasonality"])["Demand"].mean().round(2)
```

```python
df.groupby(["Promotion"])["Demand"].mean().round(2)
```

```python
pd.pivot_table(data=df, values="Demand", index="Month", columns="Category", aggfunc="mean").round(2)
```

```python
skewness = df["Demand"].skew().round(2)
kurtosis = df["Demand"].kurt().round(2)

print("Demand Skewness: ",skewness)
print("Demand kurtosis: ",kurtosis)
```

```python
sns.histplot(df["Demand"], bins=20, kde=True)
plt.title("Demand Distribution")
plt.show()
```

```python
sns.scatterplot(data=df, x = "Inventory Level", y = "Units Sold")
plt.title("Inventory vs Units Sold")
plt.show()
```

```python
sns.boxplot(data=df, x = "Category", y = "Demand")
plt.xticks(rotation=45)
plt.title("Demand by Category")
plt.show()
```

```python
sns.boxplot(data=df, x = "Weather Condition", y = "Demand")
plt.xticks(rotation=45)
plt.title("Demand by Weather Condition")
plt.show()
```

```python
monthly_demand = df.groupby("Month")["Demand"].mean().round(2)
monthly_demand
```

```python
monthly_demand.plot(kind="bar")
plt.title("Average Demand by Month")
plt.show()
```

```python
daily_demand = df.groupby("Date")["Demand"].sum().round(2)
daily_demand
```

```python
plt.figure(figsize=(10, 7))
daily_demand.plot()
plt.title("Total Daily Demand Over Time")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```

```python
sns.barplot(data=df, x="Promotion", y="Demand")
plt.title("Promotion Impact on Demand")
plt.show()
```

```python
sns.scatterplot(data=df, x="Discounted Price", y="Demand")
plt.title("Discounted Impact on Demand")
plt.show()
```

```python
df.groupby("Seasonality")["Demand"].mean().plot(kind="bar", title="Demand by Seasonality")
plt.show()
```

```python
df.groupby("Epidemic")["Demand"].mean().plot(kind="bar", title="Epidemic Impact on Seasonality")
plt.show()
```
