```python
df = pd.read_csv('demand_forecasting.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df["Weekday"] = df['Date'].dt.weekday
```

```python
df.head()
```

```python
df.columns
```

```python
df["Discounted Price"] = df['Price']*(1 - df['Discount'] / 100)
```

```python
df.head()
```

```python
df["Sell Through Rate"] = df['Units Sold'] / df['Inventory Level']
```

```python
df["Sell Through Rate"].describe()
```

```python
df.groupby('Category')["Demand"].agg(["mean",'sum',"std",]).sort_values(by = 'sum', ascending=False).round(2)
```

```python
df.groupby(["Region", "Seasonality"])["Demand"].mean().round(2)
```

```python
df.groupby(["Promotion"])["Demand"].mean().round(2)
```

```python
pd.pivot_table(data=df, values="Demand", index="Month", columns="Category", aggfunc="mean").round(2)
```
